```python
import torchvision.models
import torch
import tensorrt as trt
from torch2trt import torch2trt
```

```python
model = torchvision.models.resnet18(pretrained=True).cuda().eval().half()

input = torch.ones((1, 3, 224, 224)).cuda().half()

model_trt = torch2trt(model, [input], fp16_mode=True)
```

```python
import time
input = input.half()

model = model.half()
output = model(input)
print(output.flatten()[0:10])

t0 = time.time()
with torch.no_grad():
    for i in range(50):
        output = model(input)
        #output = model_trt(input)
t1 = time.time()

print((t1 - t0) / 50.0)
```

```python
import time
input = input.float()
output = model_trt(input)
print(output.flatten()[0:10])

t0 = time.time()
with torch.no_grad():
    for i in range(50):
        output = model_trt(input)
t1 = time.time()

print((t1 - t0) / 50.0)
```
